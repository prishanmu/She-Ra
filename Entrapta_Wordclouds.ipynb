**Notebook Purpose** : Create Wordclouds of Entrapta's spoken lines in the shape of Emily and Entrapta.

```
import warnings
warnings.filterwarnings("ignore")
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import re
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator 
from collections import Counter
import nltk
import string
import itertools
import collections
from os import path
from PIL import Image
```

```
df = pd.read_csv("she_ra_dialogue.csv")
df = df[df.dialogue!=' ']
df.dialogue.astype('str')
df.head()
```

```
#just entrapta's lines
entrap = df[df.character == 'Entrapta']
entrapta_text = str(entrap.dialogue.values)
entrapta_text = entrapta_text.lower() 
```

```
# set up Entrapta wordcloud
stopwords = set(STOPWORDS)
stopwords.add('oh')

mask = np.array(Image.open(path.join("/content/entrapta.jpg")))


wc = WordCloud(background_color="white", max_words=1000, stopwords=stopwords, mask = mask)
```

```
image_colors = ImageColorGenerator(mask)

wordcloud = wc.generate(entrapta_text)

plt.figure(figsize=(8,10))

plt.imshow(wordcloud)
plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
plt.show()
```
